```python
import pandas as pd
from sklearn.cluster import KMeans

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
d = pd.read_csv('data.csv')
```

```python
d.head()
```

```python
# Kmeans Clustring
kmeans = KMeans(n_clusters=2)
kmeans.fit(d)
y_kmeans = kmeans.predict(d)
```

```python
kmeans.cluster_centers_.shape
```

```python
# y_kmeans
```

```python
kmeans.cluster_centers_.shape
```

```python

# Plot the clustered data
plt.scatter(d.iloc[y_kmeans == 0, 0], d.iloc[y_kmeans == 0, 1], 
            s = 10, c = 'red', label = 'Type 1')
plt.scatter(d.iloc[y_kmeans == 1, 0], d.iloc[y_kmeans == 1, 1], 
            s = 10, c = 'blue', label = 'Type 2')



        
# Plotting the centroids of the clusters
# plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
#             s = 100, c = 'yellow', label = 'Centroids')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,0], c='yellow',s=100, label='Centroid')
# plt.scatter(kmeans.cluster_centers_[:,0][0],kmeans.cluster_centers_[:,0][1], c='yellow',s=100, label='Centroid')
# plt.scatter(kmeans.cluster_centers_[:,1][0],kmeans.cluster_centers_[:,1][1], c='yellow',s=100, label='Centroid')
plt.legend()
```
